# Literature on Forecasting


### Forecasting Truisms

- **Forecasts are always wrong**
- **Aggregated forecasts - This could be aggregating by SKU (Common parts) / by time (month better than days) / By Location**
- **Shorter horizon forecasts are more accurate (postponement)**

### Methods of forecasting

- Judgmental , Experimental , Causal (regression) , Time Series 

### Metrics for measuring forecasts

- Accuracy - Closeness to actual observations (MAD, MSE, RMSE, MAPE)
- Bias - Persistent tendency to over or under predict (actual-forecast)/actual+forecast)

## Understanding Forecasting Levels

Forecasting can be done at different levels of aggregations (Time) or different levels of hierarchy (Company level / region level / city level)

- Strategic 

    Horizon : Years
    
    Purposes : Capacity Planning, Business Planning 
    
    Superstore : Aggregate forecasting by category at a company level
    
    
- Tactical

    Horizon : Quarter
    
    Purposes : Sales Planning, Budgeting
    
    Superstore : Aggregate forecasting by category at a region level at a company level
    
    
- Operational 

    Horizon : Days
    
    Purposes : Production Planning, Transportation Planning
    
    Superstore : Aggregate forecasting by product at a region level at a company level

### Forecasting at a category level at a monthly frequency

```python
Technology = pd.DataFrame(data[data['Category']=='Technology']['Sales'])
Technology = Technology.resample(rule='MS').sum()
Technology.sum()
```

```python
Technology['Sales'].plot(figsize=(15,6)).autoscale(axis='x', tight=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

Based on this chart, it looks like the trend in the later days is increasing at a linear rate especially after Jul 2012 hence the data in the later part is more important than the earlier

```python
result = seasonal_decompose(Technology['Sales'], model='add')
fig = result.plot()   
fig.set_size_inches(12, 8)
```

# SMA
## Simple Moving Average

```python
Technology['Sales: 6Mth Mean'] = Technology['Sales'].rolling(window=6).mean()
Technology['Sales: 12Mth Mean'] = Technology['Sales'].rolling(window=12).mean()
Technology[['Sales','Sales: 6Mth Mean','Sales: 12Mth Mean']].plot(figsize=(14,5)).autoscale(axis='x',tight=True)
```

___
# EWMA
## Exponentially Weighted Moving Average 

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
span = 3
alpha = 2/(span+1)
print('Span : {} | Alpha : {}'.format(span, alpha))
```

```python
Technology['Sales: EWMA3']=SimpleExpSmoothing(Technology['Sales']).fit(smoothing_level=alpha,optimized=False).fittedvalues.shift(-1)
Technology[['Sales','Sales: EWMA3']].plot(figsize=(14,6)).autoscale(axis='x',tight=True);
```

```python
Technology[['Sales','Sales: 6Mth Mean','Sales: 12Mth Mean','Sales: EWMA3']].plot(figsize=(14,6)).autoscale(axis='x',tight=True)
```

___
## Double Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
Technology['Sales: DESadd03'] = ExponentialSmoothing(Technology['Sales'], trend='add').fit().fittedvalues.shift(-1)
Technology.head(15)
```

```python
Technology[['Sales','Sales: DESadd03']].plot(figsize=(14,6)).autoscale(axis='x',tight=True)
```

___
## Triple Exponential Smoothing
Triple Exponential Smoothing, the method most closely associated with Holt-Winters, adds support for both trends and seasonality in the data. 


```python
Technology['Sales: TESadd03'] = ExponentialSmoothing(Technology['Sales'], trend='add',seasonal='mul',seasonal_periods=3).fit().fittedvalues.shift(-1)
Technology.head(15)
```

```python
Technology[['Sales','Sales: TESadd03']].plot(figsize=(14,6)).autoscale(axis='x',tight=True)
```
